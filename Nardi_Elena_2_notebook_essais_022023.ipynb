```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import QuantileTransformer, PowerTransformer
from sklearn.cluster import AgglomerativeClustering, Birch, DBSCAN, KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score, calinski_harabasz_score
from yellowbrick.cluster import SilhouetteVisualizer

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import squarify
```

```python
def scores(data, estimator, min_k, max_k, threshold=0.1):
    """Cette fonction prend en entrée des données (dataframe ou matrice),
    un estimator - K-Means, Birch ou AgglomerativeClustering,
    un nombre minimum de clusters (min_k),
    un nombre maximal de clusters (max_k)
    et un valeur de seuil, threshold (que pour l'estimator Birch).
    Pour chaque k dans entre min_k et max_k (extrêmes inclus)
    elle effectue une segmentation à l'aide de l'estimator choisi
    et calcule le coefficient de silhouette, l'indice de Calinski-Harabasz
    et l'indice de Davies-Bouldin.
    La fonction retourne un dictionnaire avec tous les scores"""
    sil_scores = []
    calinski_scores = []
    davies_scores = []
    index = range(min_k, max_k+1)
    for i in index:
        if estimator == 'K-Means':
            model = KMeans(n_clusters=i, random_state=42)
        elif estimator == 'Birch':
            model = Birch(n_clusters=i, threshold=threshold)
        elif estimator == 'AHC':
            model = AgglomerativeClustering(n_clusters=i, linkage='single')
        labels = model.fit_predict(data)
        silhouette = silhouette_score(data, labels)
        CH = calinski_harabasz_score(data, labels)
        DB = davies_bouldin_score(data, labels)
        sil_scores.append(silhouette)
        calinski_scores.append(CH)
        davies_scores.append(DB)
        print('Number of clusters: ' + str(i))
        print("Silhouette Coefficient: {0:.3f}".format(silhouette))
        print("Calinski-Harabasz Index: {0:.3f}".format(CH))
        print("Davies-Bouldin Index: {0:.3f}".format(DB))
        print("--------------------------------------------------------")
    return {'Silhouette score': sil_scores,
            'Indice de Calinski-Harabasz': calinski_scores,
            'Indice de Davies-Bouldin': davies_scores}
```

```python
def DBSCAN_scores(data, labels):
    """Cette fonction prend en entrée des données (dataframe ou matrice)
    et des labels des clusters.
    La fonction print le coefficient de silhouette,
    l'indice de Calinski-Harabasz,
    l'indice de Davies-Bouldin
    et le nombre des clusters"""
    silhouette = silhouette_score(data, labels)
    CH = calinski_harabasz_score(data, labels)
    DB = davies_bouldin_score(data, labels)
    print("Number of clusters: " + str(
        len(set(labels)) - (1 if -1 in labels else 0)))
    print("Silhouette Coefficient: {0:.3f}".format(silhouette))
    print("Calinski-Harabasz Index: {0:.3f}".format(CH))
    print("Davies-Bouldin Index: {0:.3f}".format(DB))
    print("--------------------------------------------------------")
```

```python
def plot_scores(scores):
    """Cette fonction prend en entrée un dictionnaire de scores
    et fait un plot du score en fonction du nombre de clusters
    pour chaque type de score"""
    fig, ax = plt.subplots(2, 2, figsize=(15, 10))
    count = 1
    for score in scores.keys():
        plt.subplot(2, 2, count)
        plt.title(score)
        plt.plot(range(2, 10), scores[score])
        plt.xlabel('Nombre de clusters')
        count += 1
    fig.delaxes(ax[1][1])
    plt.tight_layout()
```

```python
def barplot_clusters(data):
    """Cette fonction prend en entrée un dataframe
    qui contient la variable `labels`
    - les clusters trouvés par la segmentation -
    et crée un barplot annoté avec le nombre des clients
    dans chaque cluster"""
    data = data.groupby('labels').count().sort_values(
        'recency', ascending=False
    )
    sns.barplot(data=data, x=data.index, y='recency', order=data.index)
    for i in range(data.shape[0]):
        plt.annotate(str(data.iloc[i, 0]),
                     xy=(i, data.iloc[i, 0]),
                     ha='center',
                     va='bottom')
    plt.ylabel('Nombre de clients')
    plt.xlabel('Labels')
```

```python
# paramètres matplotlib
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['axes.titlesize'] = 12
plt.rcParams['axes.labelsize'] = 11
sns.set_palette('bright')
plt.rcParams['font.size'] = 15
```

## 1.1 RFM Manuelle

Afin de construire une segmentation RFM manuellement, je vais attribuer un score aux variables `recency`, `frequency` et `payment_value` et ensuite calcule le score moyen. Pour `recency` et `payment_value` le score va aller de 1 à 5: plus le score sera élevé, plus la récence sera petite la valeur élevée. Le score sera attribué sur la base des quantiles.

Pour la fréquence, étant donné que la plupart des clients ont effectué une seule commande, le score sera de 1 ou 2: 1 si le client a commandé une seule fois, 2 sinon. Ainsi, lors du calcul du score moyen, la fréquence va automatiquement peser moins par rapport à récence et à la valeur monétaire.

```python
# dataframe pour RFM
data_RFM_manual = data[['recency', 'frequency', 'payment_value']]
```

```python
# creation des scores sur la base des quantiles
for col in ['recency', 'payment_value']:
    bins = np.array([0.0,
                     data_RFM_manual[col].quantile(0.2),
                     data_RFM_manual[col].quantile(0.4),
                     data_RFM_manual[col].quantile(0.6),
                     data_RFM_manual[col].quantile(0.8)])
    data_RFM_manual[col + '_score'] = np.digitize(data_RFM_manual[col], bins)

# inversion des notes pour la récence pour avoir les score le plus élevés
# correspondant aux récences les plus petites
data_RFM_manual['recency_score'] = 6 - data_RFM_manual['recency_score']
```

```python
# attribution du score 2 aux clients ayant effectué plusieurs commandes
data_RFM_manual['frequency_score'] = data_RFM_manual['frequency']
data_RFM_manual[data_RFM_manual['frequency_score'] > 1] = 2
```

```python
# score moyen
data_RFM_manual['score'] = data_RFM_manual[
    ['recency_score',
     'frequency_score',
     'payment_value_score']].mean(axis=1)
```

```python
# distribution des scores
plt.hist(data_RFM_manual['score'], bins=11)
plt.title('Distribution du score RFM')
plt.xlabel('Score')
plt.ylabel('Nombre de clients')
```

On pourrait distinguer les clients en trois groupes:
 -   **Worst customers**: score < 1.75
 -   **Mean customers**: 1.75 < score < 2.75
  -  **Best customers** score > 2.75

Je vais attribuer ces labels aux clients et examiner les valeurs moyens des variables pour chaque groupe.

```python
# attribution labels aux clients
data_RFM_manual['cluster'] = 'Mean customers'
data_RFM_manual.loc[data_RFM_manual['score'] < 1.75,
                    'cluster'] = 'Worst customers'
data_RFM_manual.loc[data_RFM_manual['score'] > 2.75,
                    'cluster'] = 'Best customers'
```

```python
# dataframe des valeurs moyens
data_RFM_manual_mean = data_RFM_manual.groupby('cluster').agg({
    'frequency': 'mean',
    'recency': 'mean',
    'payment_value': 'mean',
    'score': 'count'
})
data_RFM_manual_mean.rename(columns={'score': 'count'}, inplace=True)
```

Les **Worst customers** (25119 clients) sont ceux dont la dernière commande remonte en moyenne à 11 mois et ont généré en moyenne la plus petite valeurs monétaire.

Les **Mean customers** se situent entre les deux en termes de récence et valeurs monétaire. Cependant, la fréquence moyenne des 'Mean customers' est > 1, c'est à dire que ce cluster comprend les clients qui ont effectué plusieurs commandes. ### 1.2.1 RFM

#### A. Tests sur le dataset reduit

```python
# scores de référence
data_Kmeans_RFM = data_reduced[['recency', 'frequency', 'payment_value']]

scaler = MinMaxScaler()
data_Kmeans_RFM = scaler.fit_transform(data_Kmeans_RFM)

scores(data_Kmeans_RFM, 'K-Means', 2, 6)
```

```python
# log recency
data_Kmeans_RFM = data_reduced[['recency', 'frequency', 'payment_value']]

data_Kmeans_RFM['recency'] = np.log1p(data_Kmeans_RFM['recency'])

scaler = MinMaxScaler()
data_Kmeans_RFM = scaler.fit_transform(data_Kmeans_RFM)

scores(data_Kmeans_RFM, 'K-Means', 2, 6)
```

Ici, par exemple, on voit que la transformation logarithme de la variable `recency` perme une amélioration des scores.

```python
# log payment_value
data_Kmeans_RFM = data_reduced[['recency', 'frequency', 'payment_value']]

data_Kmeans_RFM['payment_value'] = np.log1p(data_Kmeans_RFM['payment_value'])

scaler = MinMaxScaler()
data_Kmeans_RFM = scaler.fit_transform(data_Kmeans_RFM)

scores(data_Kmeans_RFM, 'K-Means', 2, 6)
```

Par contre, la transformation logarithme de la variable `payment_value` empire les scores par rapport aux scores de référence (sans transformation des variables).

```python
# quantile transformation recency et frequency
data_Kmeans_RFM = data_reduced[['recency', 'frequency', 'payment_value']]

qt = QuantileTransformer(output_distribution='normal')
data_Kmeans_RFM[['recency', 'frequency']] = qt.fit_transform(
    data_Kmeans_RFM[['recency', 'frequency']])

scaler = MinMaxScaler()
data_Kmeans_RFM = scaler.fit_transform(data_Kmeans_RFM)

scores(data_Kmeans_RFM, 'K-Means', 2, 6)
```

Le modèle entrainé en utilisant la transformation quantile de `recency` et `frequency` est celui qui donne les meilleurs scores et je vais l'entrainer sur le dataset entier.

Pour tous les modèles non supervisés, les meilleurs résultats ont été obtenus avec le seul `review_score` en plus des variables RFM et avec la transformation quantile des variables `recency`, `frequency` et `review_score`.

## 2.3 Suggestions d'actions marketing

A partir des groupes clients identifiés, on peut réfléchir à des stratégies marketing adaptées, par exemple:

- **Recent customers**: ils ont été assez satisfaits du produit, il faut s’assurer qu’il reviennent. On pourrait leur envoyer un message de remerciement pour l’achat, leur demander l’avis sur l’experience d’achat et comment l’améliorer. On pourrait aussi offrir un cadeau aux clients qui n’ont pas été très satisfaits du produit acheté (réduction, coupons, etc.)
- **Best customers**: il s’agit de client qui ont commandé plusieurs fois et sont assez satisfaits de produits. Il faut continuer à les motiver, par exemple avec l’implementation de programmes de fidélité
- **One-timers**: ils ont été assez satisfait du produit. Il ne faut pas qu’ils oublient Olist, on pourrait réaliser une campagne de mails personnalisés en proposant des produits supplémentaires qui complètent leur achat, ou leur proposer des articles exclusifs 
- **Unhappy customers**: il faudrait essayer de le recontacter en leur proposant les nouveaux produit dans le catalogue, en incluant les avis positifs sur ces produits, et leur proposer des promotions.
